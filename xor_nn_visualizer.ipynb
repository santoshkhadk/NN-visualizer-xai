```python



import numpy as np


X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

Y = np.array([[0],
              [1],
              [1],
              [0]])

print("XOR Inputs:\n", X)
print("XOR Outputs:\n", Y)



np.random.seed(42)
input_size = 2
hidden_size = 4
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))




def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return (Z > 0).astype(float)

def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))




learning_rate = 0.1
epochs = 5000




for epoch in range(epochs):
    

    Z1 = X.dot(W1) + b1
    A1 = relu(Z1)
    
    Z2 = A1.dot(W2) + b2
    A2 = sigmoid(Z2)
    
   
    loss = -np.mean(Y*np.log(A2+1e-8) + (1-Y)*np.log(1-A2+1e-8))
    
 
    dZ2 = A2 - Y
    dW2 = A1.T.dot(dZ2)
    db2 = np.sum(dZ2, axis=0, keepdims=True)
    
    dA1 = dZ2.dot(W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = X.T.dot(dZ1)
    db1 = np.sum(dZ1, axis=0, keepdims=True)
    
    
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")




print("\nFinal Predictions:")
pred = (A2 > 0.5).astype(int)
print(pred)




print("\nHidden Layer Activations (A1):")
print(A1)
print("\nOutput Layer (A2):")
print(pred)
```

```python


# Gradient of output w.r.t input
dA2_dZ2 = A2 * (1 - A2)  # sigmoid derivative
dZ2_dA1 = W2               # derivative of Z2 w.r.t A1
dA1_dZ1 = relu_derivative(Z1)
dZ1_dX = W1                 # derivative of Z1 w.r.t X

# Chain rule: dA2/dX = dA2/dZ2 * dZ2/dA1 * dA1/dZ1 * dZ1/dX
input_contributions = dA2_dZ2.dot(dZ2_dA1.T) * dA1_dZ1
input_contributions = input_contributions.dot(dZ1_dX.T)

print("\nFeature Contributions (how much each input affects output):")
for i, xi in enumerate(X):
    print(f"Input {xi} -> Contribution: {input_contributions[i]}")
```
